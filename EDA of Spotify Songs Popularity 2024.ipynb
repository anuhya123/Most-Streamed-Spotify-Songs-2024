# What is Exploratory Data Analysis?

Exploratory Data Analysis (EDA) is a method of analyzing datasets to understand their main characteristics. It involves summarizing data features, detecting patterns, and uncovering relationships through visual and statistical techniques. EDA helps in gaining insights and formulating hypotheses for further analysis.

In our data-driven processes, we prioritize refining our raw data through the crucial stages of EDA (Exploratory Data Analysis). Both data pre-processing and feature engineering play pivotal roles in this endeavor. EDA involves a comprehensive range of activities, including data integration, analysis, cleaning, transformation, and dimension reduction.

Data pre-processing involves cleaning and preparing raw data to facilitate feature engineering. Meanwhile, feature engineering entails employing various techniques to manipulate the data. This may include adding or removing relevant features, handling missing data, encoding variables, and dealing with categorical variables, among other tasks.

Undoubtedly, feature engineering is a critical task that significantly influences the outcome of a model. It involves crafting new features based on existing data while pre-processing primarily focuses on cleaning and organizing the data.

# Step 5: Creating Features -

We will play around with the variable Release Date in our dataset. If we see the sample data, the column shows the release date of the song/track. 

It wont help us much or becomes difficult to analyze the data. Hence, we will create new features like day, month and year for more better analysis.

```python
# Convert 'Release Date' to datetime format
df['Release Date'] = pd.to_datetime(df['Release Date'])

# Extract day, month, and year
df['Day'] = df['Release Date'].dt.day
df['Month'] = df['Release Date'].dt.month
df['Year'] = df['Release Date'].dt.year
```

```python
df
```

An ISRC code consists of 12 characters divided into four parts: <br>
CountryCode (2 characters): Represents the country where the recording was first released.<br>
Registrant Code (3 characters): Identifies the owner or administrator of the recording.<br>
Year of Release (2 characters): Indicates the year the recording was first released.<br>
Serial Number (5 characters): A unique number assigned by the registrant to the recording.<br>

Based on this, a file with all the country codes and the agency name and location of the agency is pulled into our data analysis. From this file, we will map the country code with the existing data frame and create a new column with the Country name i.e., Location of the agency.

```python
df['Country Code'] = df['ISRC'].str[:2]
df['ISRC_Registrant_Code'] = df['ISRC'].str[2:5]
df['ISRC_Year'] = df['ISRC'].str[5:7]
df['ISRC_Serial_Number'] = df['ISRC'].str[7:]
df
```

# Step 6: Data Cleaning/Wrangling -

Some names of the variables are not relevant and not easy to understand. Some data may have data entry errors, and some variables may need data type conversion. We need to fix this issue in the data.

```python
df.info()
```

```python
#My month column is represinting its values in the number form. 
#Instead, I want to change it to the month name for more easy analysis and visualization.

# Create a dictionary mapping month numbers to names
month_names = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}

# Map the month numbers to their corresponding names
df['Month'] = df['Month'].map(month_names)
```

```python
df
```

```python
#On proper observation, the year in the ISRC code does not match the year column from the release date for few records. 
#Hence, we are cleaning such data rows.

# Check if the last two digits of yyyy match yy and replace if necessary
df['Year'] = df.apply(lambda row: str(row['Year'])[:-2] + row['ISRC_Year'] if str(row['Year'])[-2:] != row['ISRC_Year'] else row['Year'], axis=1)

df.head()
```

```python
df.info()
```

```python
# Convert columns to numeric

df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df['ISRC_Year'] = pd.to_numeric(df['ISRC_Year'], errors='coerce')

df.info()
```

```python
df.shape
```

Before we do EDA, lets separate Numerical and categorical variables for easy analysis

```python
categorical = df.select_dtypes(include=['object']).columns
numerical = df.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(categorical)
print("Numerical Variables:")
print(numerical)
```

```python
# CHecking if my categorical columns have any special characters. If yes, removing those rows.

special_chars_regex = r'[^a-zA-Z0-9\s]'

for col in categorical:
    mask = df[col].astype(str).str.contains(special_chars_regex)
    df = df[~mask]

df
```

What Skew Value Indicates -

Skewness is a statistical measure that quantifies the asymmetry of a probability distribution. It tells you whether a dataset is symmetrical or skewed.

Positive Skew (Skew > 0): The distribution has a long tail to the right. This indicates that there are more smaller values and fewer larger values.

Negative Skew (Skew < 0): The distribution has a long tail to the left. This indicates that there are more larger values and fewer smaller values.

Zero Skew (Skew â‰ˆ 0): The distribution is approximately symmetrical.

In summary, the skew value provides information about the shape of your data distribution. A positive or negative skew can influence the choice of statistical methods and interpretation of results.

Log transformation can help in normalization, so this variable can maintain standard scale with other variables:

```python
sns.distplot(df['Spotify Streams'], axlabel='Spotify Streams');
```

```python
# Function for log transformation of the column
def log_transform(data,col):
    for colname in col:
        if (data[colname] == 1.0).all():
            data[colname + '_log'] = np.log(data[colname]+1)
        else:
            data[colname + '_log'] = np.log(data[colname])
    data.info()
    
log_transform(df,['Spotify Streams', 'Spotify Playlist Count'])
```

```python
#Log transformation of the feature 'Kilometers_Driven'
sns.distplot(df['Spotify Streams'], axlabel='Spotify Streams');
```
